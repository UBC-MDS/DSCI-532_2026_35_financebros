```python
import duckdb #we use duckdb to convert csv to parquet for duckdb

CSV_Parquet = ({
    "../data/close.csv" : "../data/close.parquet", 
    "../data/metric.csv" : "../data/metric.parquet", 
    "../data/spy.csv" : "../data/spy.parquet", 
    "../data/watchlist.csv" : "../data/watchlist.parquet", 
    })   

for file in CSV_Parquet:
    duckdb.execute(f"""
        COPY (SELECT * FROM read_csv_auto('{file}'))
        TO '{CSV_Parquet[file]}' (FORMAT PARQUET)
    """)
```
